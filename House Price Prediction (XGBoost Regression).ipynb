Importing dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

Importing the dataset 

```python
house_price_dataset = sklearn.datasets.fetch_california_housing()
```

```python
print(house_price_dataset)
```

```python
#Loading the dataset to a Pandas Dataframe
house_price_dataframe = pd.DataFrame(house_price_dataset.data, columns = house_price_dataset.feature_names)
```

```python
house_price_dataframe.head()
```

```python
# Adding the target(price) column to DataFrame
house_price_dataframe['price'] = house_price_dataset.target
```

```python
house_price_dataframe.head()
```

```python
# checking the number of rows and columns in the dataframe
house_price_dataframe.shape
```

```python
# check for missing values
house_price_dataframe.isnull().sum()
```

```python
# statistical measures of the dataset
house_price_dataframe.describe()
```

Undersanding the correlation between the features
1.Positive
2.Negative

```python
correlation = house_price_dataframe.corr()
```

```python
# constructing a heatmap to understand the correlation
plt.figure(figsize = (10,10))
sns.heatmap(correlation, cbar=True, square = True, fmt = '0.1f', annot = True, annot_kws={'size':8}, cmap='Blues')
```

Splitting the data and target

```python
X = house_price_dataframe.drop(['price'], axis = 1)
Y = house_price_dataframe['price']
```

```python
print(X)
print(Y)
```

Splitting the data into training and testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 2)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_norm = scaler.fit_transform(X_train)
```

```python
print(X.shape, X_train_norm.shape, X_test.shape)
```

MODEL TRAINING

XGBoost Regressor

```python
#loading the model
model = XGBRegressor()
```

```python
#training the model with X_train_norm
model.fit(X_train, Y_train)
```

Evaluation

Prediction on training data

```python
# Accuracy for the prediction on training data
train_data_prediction = model.predict(X_train)
```

```python
print(train_data_prediction)
```

```python
# R-squared error
# Y_train is the original value and train_data_prediction is our prediction
score_1 = metrics.r2_score(Y_train, train_data_prediction)

#Mean absolute error
score_2 = metrics.mean_absolute_error(Y_train, train_data_prediction)

print("R-squared error:", score_1)
print("Mean absolute error:", score_2)
```

Visualizing the actual Prices and predicted prices

```python
plt.scatter(Y_train, train_data_prediction)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual price vs Predicted price")
plt.show()
```

Prediction on Test data

```python
# Accuracy for the prediction on testing data
test_data_prediction = model.predict(X_test)
```

```python
# R-squared error
score_1 = metrics.r2_score(Y_test, test_data_prediction)

#Mean absolute error
score_2 = metrics.mean_absolute_error(Y_test, test_data_prediction)

print("R-squared error:", score_1)
print("Mean absolute error:", score_2)
```

Visualizing the actual Prices and predicted prices

```python
plt.scatter(Y_test, test_data_prediction)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual price vs Predicted price")
plt.show()
```
